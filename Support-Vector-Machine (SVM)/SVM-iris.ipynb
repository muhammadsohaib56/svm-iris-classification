```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['target_names'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
%matplotlib inline      #inline magic only jupyter notebook
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
#draw a scatter plot
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='green',marker='+')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='blue',marker='*')
```

```python
#draw a scatter plot
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='green',marker='+')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='blue',marker='*')
```

```python
X = df.drop(['target','target_names'], axis='columns')
X.head()
```

```python
Y = df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.svm import SVC
model = SVC()    #C is Regularization increases decreasing score
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[5.1,3.1,1.4,0.2]])
```

```python
import joblib
```

```python
joblib.dump(model,'SVM-model')
```
